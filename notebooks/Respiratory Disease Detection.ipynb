```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Respiratory virus detections 2023.xlsx")
df
```

```python
virus_cols = [
    'Covid_Ratio', 'Flu_Ratio', 'RSV_Ratio',
    'HPIV_Ratio', 'ADV_Ratio', 'HMPV_Ratio',
    'EV/RV_Ratio', 'HCoV_Ratio'
]

df = df[['WEEK'] + virus_cols].copy()

# % Removal → float conversion
for c in virus_cols:
    df[c] = (
        df[c].astype(str).str.replace('%', '', regex=False).replace('', '0').astype(float)
        / 100.0
    )
    
df
```

```python
df.describe()
```

```python
# Build 3-category-flag: 0 (Normal), 1 (Watch), 2 (Severe) for core variables influencing demand, supply chain directly
core_cols = ['Covid_Ratio', 'Flu_Ratio', 'RSV_Ratio'] 
  
# Threshold Creation
thresholds = {}

for col in core_cols:
    q = df[col].quantile([0.50, 0.75])
    thresholds[col] = {
        'q50': q.loc[0.50],
        'q75': q.loc[0.75]
    }


# Risk Flagging
def risk_flag(value, q50, q75):
    if value <= q50:
        return 0          # Normal
    elif value <= q75:
        return 1          # Watch
    else:
        return 2          # Severe


# Flag Column Creation
for col in core_cols:
    df[f'{col}_FLAG'] = df[col].apply(
        lambda x: risk_flag(
            x,
            thresholds[col]['q50'],
            thresholds[col]['q75']
        )
    )
    
# Use 'mean' to provide a more stable signal, reducing spikes
df['REDUCED_SUPPLY_CAPACITY_CORE'] = df[['Covid_Ratio_FLAG', 'Flu_Ratio_FLAG', 'RSV_Ratio_FLAG']].mean(axis=1)

def core_level(x):
    if x <= 0.5:
        return 0
    elif x <= 1.5:
        return 1
    else:
        return 2

df['REDUCED_SUPPLY_CAPACITY_CORE'] = df['REDUCED_SUPPLY_CAPACITY_CORE'].apply(core_level)
```

```python
# Consider additive pressure by using the rest auxiliary variables that amplify the risk
aux_cols = ['HPIV_Ratio', 'ADV_Ratio', 'HMPV_Ratio', 'EV/RV_Ratio', 'HCoV_Ratio']

# Threshold Creation
aux_thresholds = {}
for col in aux_cols:
    q = df[col].quantile([0.50, 0.75])
    aux_thresholds[col] = {
        'watch': q.loc[0.50],
        'severe': q.loc[0.75]
    }

# Risk Flagging
for col in aux_cols:
    df[f'{col}_FLAG'] = df[col].apply(
        lambda x: risk_flag(
            x,
            aux_thresholds[col]['watch'],
            aux_thresholds[col]['severe']
        )
    )

# Additive Pressure (Depending on the severe count)
df['AUX_SEVERE_COUNT'] = (
    df[[f'{c}_FLAG' for c in aux_cols]] == 2
).sum(axis=1)

def aux_pressure(n):
    if n <= 2:
        return 0
    elif n <= 4:
        return 1
    else:
        return 2

df['AUX_ADDITIVE_PRESSURE'] = df['AUX_SEVERE_COUNT'].apply(aux_pressure)

# Combine with the reduced capacity for core variables
df['REDUCED_SUPPLY_CAPACITY_FINAL'] = (df['REDUCED_SUPPLY_CAPACITY_CORE']+df['AUX_ADDITIVE_PRESSURE']).clip(upper=2)
```

```python
# Add T+2 lag; disease outbreaks(t) -> reduced supply capacity (t+2)
df = df.sort_values('WEEK').reset_index(drop=True)
df['REDUCED_SUPPLY_CAPACITY_LAG2'] = (df['REDUCED_SUPPLY_CAPACITY_FINAL'].shift(2).fillna(0).astype(int))

df
```
